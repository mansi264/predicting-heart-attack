Your total serum cholesterol consists of high-density lipoprotein, or HDL cholesterol; low-density lipoprotein, or LDL cholesterol; and triglycerides. Levels are reported in milligrams per deciliter of blood, or mg/dL. The American Heart Association defines normal total cholesterol as less than 200 mg/dL. Total serum cholesterol above 200 is "borderline high" and raises your risk of heart disease. Total serum cholesterol above 240 is considered high. A person with total cholesterol that high has double the risk of developing heart disease compared to someone whose total cholesterol is below 200.

*However, since it is usually difficult to determine exactly where the ST segment ends and the T wave begins, the relationship between the ST segment and T wave should be examined together.
* Flat, downsloping, or depressed ST segments may indicate coronary ischemia :  is a medical term for a reduced blood flow in the coronary circulation

* Left ventricular hypertrophy is thickening of the walls of the lower left heart chamber. During left ventricular hypertrophy, the thickened heart wall can become stiff. Blood pressure in the heart increases. The changes make it harder for the heart to effectively pump blood. Eventually, the heart may fail to pump with as much force as needed.

***TESTING OF DEPENDENCE OF CATEGORICAL VARIABLES ON TARGET VARIABLE***

We then perform the chi square test of independence of attributes to check whether or not our categorical variables have an impact on heart disease.

```python
cg_sex=pd.crosstab(index = cat_data['target'], columns= cat_data['Sex'])
print(cg_sex)

cg_FBP=pd.crosstab(index = cat_data['target'], columns= cat_data['FastingBS'])
print(cg_FBP)

cg_CPT=pd.crosstab(index = cat_data['target'], columns= cat_data['ChestPainType'])
print(cg_CPT)

cg_RECG=pd.crosstab(index = cat_data['target'], columns= cat_data['RestingECG'])
print(cg_RECG)

cg_EA=pd.crosstab(index = cat_data['target'], columns= cat_data['ExerciseAngina'])
print(cg_EA)

cg_ST=pd.crosstab(index = cat_data['target'], columns= cat_data['ST_Slope'])
print(cg_ST)
```

H0 : The heart disease is indendent of the categorical variables.
H1 : not HO

```python
from scipy import stats
chi2,pvalue,df,ex = stats.chi2_contingency(cg_sex)
print( chi2, pvalue ,df)
```

```python
from scipy import stats
chi2,pvalue,df,ex = stats.chi2_contingency(cg_FBP)
print( chi2, pvalue ,df)
```

```python
chi2,pvalue,df,ex = stats.chi2_contingency(cg_CPT)
print( chi2, pvalue ,df)
```

```python
chi2,pvalue,df,ex = stats.chi2_contingency(cg_RECG)
print( chi2, pvalue ,df)
```

```python
chi2,pvalue,df,ex = stats.chi2_contingency(cg_EA)
print( chi2, pvalue ,df)
```

```python
chi2,pvalue,df,ex = stats.chi2_contingency(cg_ST)
print( chi2, pvalue ,df)
```

Here , we can say that the p-value(0.00)< level of significance(0.05) for all our categorical variables.

Thus, we conclude that the heart disease is affected by our categorical variables as they are not independent.

***ONE HOT ENCODING OF TRAIN DATA***

One hot encoding is a technique that we use to represent categorical variables as numerical values in a machine learning model.



The provided code is a Python function called OneHotEncoding. This function is designed to perform one-hot encoding on a pandas DataFrame column and then append the resulting one-hot encoded columns to the DataFrame.

```python
def OneHotEncoding(dfcolumn):
  global df
  dfcolumn.nunique()
  len(df.columns)
  finallencol = (dfcolumn.nunique() - 1) + (len(df.columns)-1)
  dummies = pd.get_dummies(dfcolumn, drop_first=True, prefix=dfcolumn.name)
  df=pd.concat([df,dummies],axis='columns')
  df.drop(columns=dfcolumn.name,axis=1,inplace=True)
  if(finallencol==len(df.columns)):
    print('OneHotEncoding is sucessfull')
    print('')
  else:
    print('Unsucessfull')
  return df.head(5)
```

```python
df = tr_data
OneHotEncoding(df['ChestPainType'])
OneHotEncoding(df['Sex'])
OneHotEncoding(df['RestingECG'])
OneHotEncoding(df['ExerciseAngina'])
OneHotEncoding(df['ST_Slope'])
tr_data=df
```

***IMPUTING MISSING VALUES OF CHOLESTEROL IN TRAIN DATA***

The missing values in the 'Cholesterol' column of 'tr_data' have been imputed using the k-nearest neighbors algorithm, and the column has been updated with the imputed values. Note that the original 'Cholesterol' column was modified in place, and the code assumes that there are no missing values in other columns of the DataFrame 'tr_data'.






```python
from sklearn.impute import KNNImputer
tr_data['Cholesterol'].replace(to_replace = 0, value =np.nan, inplace=True)
KNN_imputed = KNNImputer(n_neighbors=5)
I=KNN_imputed.fit_transform(tr_data)
Cholesterol=[]
for i in range(0,len(tr_data)):
  Cholesterol.append(I[i][2])
tr_data['Cholesterol']=Cholesterol
```

Thus we see that now there are no missing values in our choelstrol variable.

***FITTING OF LOGISTIC REGRESSION***

The following code shows how to perform logistic regression using scikit-learn's LogisticRegression class for a binary classification task.

```python
from sklearn.linear_model import LogisticRegression
y_train = tr_data["target"]
x_train = tr_data.drop("target",axis=1)
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

The following code shows how to use the trained logistic regression model (lr) to make predictions on a test dataset (t_data).

```python
y_test = t_data["target"]
x_test = t_data.drop("target",axis=1)
pred = lr.predict(x_test)
```

The following code shows how to calculate the accuracy of the logistic regression model's predictions on the test dataset using scikit-learn's accuracy_score function.

***MODEL ACCURACY***

```python
from sklearn.metrics import accuracy_score, confusion_matrix
#Accuracy on test data
accuracy_score(y_test, pred)
```

The following code calculates the accuracy of the logistic regression model's predictions on the training dataset (x_train, y_train) using scikit-learn's accuracy_score function.

```python
#accuracy on train data
accuracy_score(y_train, lr.predict(x_train))
```

The above code snippet shows the accuracy scores of a logistic regression model on both the test and training datasets:

accuracy_score(y_test, pred): The accuracy score on the test data is approximately 0.8641, which means the logistic regression model correctly predicted around 86.41% of the instances in the test dataset.

accuracy_score(y_train, lr.predict(x_train)): The accuracy score on the training data is approximately 0.8733, which means the logistic regression model correctly predicted around 87.33% of the instances in the training dataset.

It's important to note that the accuracy on the training data (87.33%) is slightly higher than the accuracy on the test data (86.41%). This is not uncommon, as the model is typically optimized to perform well on the training data during the training process. However, the true test of a model's performance is its ability to generalize to new, unseen data (i.e., the test data).

A small difference between the training and test accuracies is generally acceptable
